```python
numeric = df.columns[:-1]
```

```python
# provera autlajera
fig, ax = plt.subplots(8, 2, figsize=(15, 25))
for var, subplot in zip(numeric, ax.flatten()):
    sns.boxplot(x=df['Class'], y=df[var], ax=subplot)
```

```python
# uklanjanje outlajera - TODO :(
for var in numeric:
    IQR = df[var].quantile(.75) - df[var].quantile(.25)
    LL = df[var].quantile(.25) - 1.5 * IQR
    UL = df[var].quantile(.75) + 1.5 * IQR
    df[var] = np.where(df[var] < LL, LL, df[var])
    df[var] = np.where(df[var] > UL, UL, df[var])
```

```python
# provera autlajera - sada ne treba da ih bude
fig, ax = plt.subplots(8, 2, figsize=(15, 25))
for var, subplot in zip(numeric, ax.flatten()):
    sns.boxplot(x=df['Class'], y=df[var], ax=subplot)
```
